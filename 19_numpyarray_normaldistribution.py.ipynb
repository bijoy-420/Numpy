```python
#Normal Distribution (gauessian Distribution)-very important.
#Random.normal() method-loc,scale,size

from numpy import random
shared=random.normal(size=(2,3))
print(shared)

```

```python
#Normal Distribution (gauessian Distribution)-very important.
#Random.normal() method-loc(mean),scale(standard deviation),size()

#Here we will  generate a randdom normal dist of size 2*3 with meann at 1 and standard deviation of 2;
from numpy import random
shared=random.normal(loc=1,scale=2,size=(2,3))
print(shared)

```

```python
#visualization of normal distribution 
#this curve of a normal distribution is also called bell curve

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.normal(size=1000),hist=False)
plt.show()
```
