```python
#Rayleigh dist-it is used in signal processing
#parameters-sclae(std,how flat the distribution is...default 1.0),size

#sample for R1 with scale  of 2.0 with size of 2
from numpy import random
shared=random.rayleigh(scale=2,size=(2,3))
print(shared)
```

```python
#visualization of the rayleigh dist
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.rayleigh(size=1000),hist=False)
plt.show()
```
