```python
#chi square dist-it is basically use as basis to verify the hypothesis
#parameters-df(degree of freesom),size
#sample for chi squared dist with df 2 with size 2*3

from numpy import random 
shared=random.chisquare(df=2,size=(2,3))
print(shared)
```

```python
#visualization of chisquare dist
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.chisquare(df=1,size=1000),hist=False)
plt.show()
```
